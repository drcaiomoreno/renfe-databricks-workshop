Prompt: Write a Spark program that reads the data from this CSV file, using ; as the delimiter, and displays its contents in a clear, readable format.

URL = /Volumes/classic_stable_7807mj_catalog/renfe/renfe_csv/renfe_estaciones_csv.csv

```python
df = spark.read.option("header", "true").option("inferSchema", "true").option("delimiter", ";").csv("/Volumes/workspace/renfe/renfe_csv/renfe_estaciones_csv.csv")
display(df)
```

Write to a Delta Table inside Unity Catalog

```python
df.write.format("delta").mode("overwrite").saveAsTable("workspace.renfe.renfe_estaciones_delta")
```
